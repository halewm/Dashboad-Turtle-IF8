### Cleaning Data

Mengubah tipe data `dteday` dari `object` menjadi `datetime`

```python
day_df['dteday'] = pd.to_datetime(day_df['dteday'])
day_df.info()
```

## Exploratory Data Analysis (EDA)

### Pertanyaan 1 : Bagaimana tren penggunaan sepeda setiap bulannya pada tahun 2011? - (NIM-NAMA)

```python
day_df_2011 = day_df[day_df['dteday'].dt.year == 2011]
day_df_2011
```

```python
month_df_2011 = day_df_2011['mnth'].unique()
month_df_2011
```

Menghitung rata-rata penggunaan sepeda setiap bulannya pada tahun 2011

```python
rataRata_2011 = day_df_2011.groupby('mnth')['cnt'].mean()
rataRata_2011
```

### Pertanyaan 2 : Dari tren penggunaan sepeda pada tahun 2011, faktor apa yang mempegaruhi penggunaan sepeda setiap bulannya pada tahun 2011? - (NIM-NAMA)

```python
rataRata_cuaca_2011 = day_df_2011.groupby('mnth')['temp'].mean()
rataRata_cuaca_2011
```

```python
rataRata_hum_2011 = day_df_2011.groupby('mnth')['hum'].mean()
rataRata_hum_2011
```

```python
merged_df = pd.merge(rataRata_2011, rataRata_cuaca_2011, on='mnth', suffixes=('_cnt', '_temp'))
merged_df_2011 = pd.merge(merged_df, rataRata_hum_2011, on='mnth')
merged_df_2011
```

### Pertanyaan 3 & 4: Bagaimana tren penggunaan sepeda setiap bulannya pada tahun 2012? - (NIM-NAMA)

```python
day_df_2012 = day_df[day_df['dteday'].dt.year == 2012]
day_df_2012
```

```python
month_df_2012 = day_df_2012['mnth'].unique()
month_df_2012
```

```python
rataRata_2012 = day_df_2012.groupby('mnth')['cnt'].mean()
rataRata_2012
```

```python
rataRata_cuaca_2012 = day_df_2012.groupby('mnth')['temp'].mean()
rataRata_cuaca_2012
```

## Visualization & Explanatory Analysis

### Pertanyaan 1 : Bagaimana tren penggunaan sepeda setiap bulannya pada tahun 2011? - (NIM-NAMA)

```python
plt.figure(figsize=(10, 6))
plt.plot(month_df_2011, rataRata_2011, marker='o', label='Rata-rata Penggunaan Sepeda')
plt.title('Rata-rata Penggunaan Sepeda Setiap Bulan pada Tahun 2011')
plt.xlabel('Bulan')
plt.ylabel('Rata-rata Jumlah Sepeda Digunakan')
plt.xticks(month_df_2011, ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember'], rotation=90)
plt.legend()
plt.grid(True)
plt.show()
```

### Pertanyaan 2 : Dari tren penggunaan sepeda pada tahun 2011, faktor apa yang mempegaruhi penggunaan sepeda setiap bulannya pada tahun 2011? - (NIM-NAMA)

```python
plt.figure(figsize=(10, 6))
plt.plot(month_df_2011, rataRata_cuaca_2011, marker='o', label='Suhu rata-rata')
plt.title('Suhu rata-rata setiap bulannya pada tahun 2011')
plt.xlabel('Bulan')
plt.ylabel('Rata-rata Jumlah Sepeda Digunakan')
plt.xticks(month_df_2011, ['Januari', 'Februari', 'Maret', 'April', 'Mei', 'Juni', 'Juli', 'Agustus', 'September', 'Oktober', 'November', 'Desember'], rotation=90)
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(22, 8))
sns.barplot(x='mnth', y='cnt', hue='weathersit', data=day_df_2011, errorbar=None)

legend_labels = {
    1: 'Clear, Few clouds, Partly cloudy, Partly cloudy',
    2: 'Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist',
    3: 'Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds',
    4: 'Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog'
}
plt.legend(title='Weather Situation', loc='upper right', bbox_to_anchor=(1.2, 1), labels=[legend_labels[i] for i in sorted(legend_labels)])
plt.title('Jumlah total sepeda berdasarkan situasi cuaca (weather situation) setiap bulannya (2011)')
plt.xlabel('Month')
plt.ylabel('Total Count')
plt.show()
```

```python
correlation_matrix_2011 = merged_df_2011.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_2011, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap: Suhu, Kelembaban, dan Penggunaan Sepeda')
plt.show()
```

### Pertanyaan 4 : Dari tren penggunaan sepeda pada tahun 2012, faktor apa yang mempegaruhi penggunaan sepeda setiap bulannya pada tahun 2012? - (NIM-NAMA)
